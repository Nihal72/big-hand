```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train = pd.read_csv(r"../Data/RawData/train.csv")
test = pd.read_csv(r"../Data/RawData/test.csv")
```

```python
## First drop these columns which are useless for us
## 1. PassengerID - Simple RowNumber
## 2. Cabin - Too many missing values
## 3. Ticket - Dubplicate values
```

```python
train.drop(["PassengerId", "Cabin", "Ticket"] , axis = 1 , inplace = True)
test.drop(["PassengerId", "Cabin", "Ticket"] , axis = 1 , inplace = True)
```

```python
## Now let's going forward lets kick some by making new features.
## let's take a look how Name can give us information
```

```python
train_title = list(map(lambda x : x.split(".")[0].split(" ")[1] , train["Name"]))
```

```python
train["Title"] = train_title
```

```python
## let's refine the title there are too many rare title.
```

```python

for dataset in combine:
    dataset['Title'] = dataset['Title'].replace(['Lady', 'Countess','Capt', 'Col',\
     'Don', 'Dr', 'Major', 'Rev', 'Sir', 'Jonkheer', 'Dona'], 'Rare')

    dataset['Title'] = dataset['Title'].replace('Mlle', 'Miss')
    dataset['Title'] = dataset['Title'].replace('Ms', 'Miss')
    dataset['Title'] = dataset['Title'].replace('Mme', 'Mrs')
    
# combine[0][['Title', 'Survived']].groupby(['Title'], as_index=False).mean()
```

```python
combine[0].groupby("Title" , as_index = False)["Survived"].mean()
```

```python
pd.crosstab(train["Title"] , train["Sex"])
```

```python
## convert Titles into oridnal data
```

```python
title_mapping = {"Mr": 1, "Miss": 2, "Mrs": 3, "Master": 4, "Rare": 5}
for dataset in combine:
    dataset['Title'] = dataset['Title'].map(title_mapping)
    dataset['Title'] = dataset['Title'].fillna(0)

train.head()
```

```python
## now we can drop Name from the datasets
```

```python
for dataset in combine:
    dataset.drop(["Name"] , axis = 1 , inplace = True)
    
```

```python
for dataset in combine:
    dataset["Sex"] = dataset["Sex"].map({"male":1 , "female":0})
```

```python
guess_age = np.zeros((2, 3))
```

```python
for i in range(0,2):
    for j in range(0, 3):
        for dataset in combine:
            age_g = dataset[(dataset["Sex"] == i) & (dataset["Pclass"] == j+1)]["Age"].median()
            
            guess_age[i, j] = age_g
            
            dataset.loc[(dataset["Age"].isna()) & (dataset["Sex"] == i) & (dataset["Pclass"] == j + 1) , "Age"] = age_g
#             dataset["Age"] = dataset["Age"].astype(int)
```

```python
## let's create band of Age
```

```python
for dataset in combine:
    dataset["Ageband"] = pd.cut(dataset["Age"] , 5)
```

```python
train.head()
```

```python
## lets check is Ageband really helpful for predicting survival
```

```python
train.groupby("Ageband" , as_index=False)["Survived"].mean()
```

```python
# Yes this is giving us huge information .
# 1. persons age below 16 are more likely to be surviv.
# 2. person above 65 survival rate is very low.
```

```python
## now lets convert Age feature into ordinal features.
```

```python
for dataset in combine:
    dataset.loc[dataset["Age"] <= 16 , "Age"] = 0
    dataset.loc[(dataset["Age"] > 16) & (dataset["Age"] <= 32) , "Age"] = 1
    dataset.loc[(dataset["Age"] > 32) & (dataset["Age"] <= 48) , "Age"] = 2
    dataset.loc[(dataset["Age"] > 48) & (dataset["Age"] <= 64) , "Age"] = 3
    dataset.loc[(dataset["Age"] > 64)  , "Age"] = 4
```

```python
train["Age"].value_counts()
```

```python
# Now we can remove Ageband
```

```python
for dataset in combine:
    dataset.drop(["Ageband"] , axis = 1 , inplace = True)
```

```python
train.head()
```

```python
## now its time to create new feature
# FamilySize = SibSp + Parch

for dataset in combine:
    dataset["FamilySize"] = dataset["SibSp"] + dataset["Parch"] + 1
```

```python
train.groupby("FamilySize" , as_index = False)["Survived"].mean().sort_values(by = "Survived" , ascending = False)
```

```python
train[train["FamilySize"] == 11]
```

```python
## This information is very usefull for, bigger the family size lesser survival chance, FamilySize > 7 less chance
## but for 4 ,3 ,2 chaces are great, but it will confuse our Logistic regressoin model, so for that let's create 
## different dataset for that also.
```

```python
## new features IsAlone

for dataset in combine:
    dataset["IsAlone"] = 0
    dataset.loc[dataset["FamilySize"] == 1 , "IsAlone"] = 1
```

```python
train.groupby("IsAlone" , as_index = False)["Survived"].mean().sort_values(by = "Survived" , ascending = False)
```
